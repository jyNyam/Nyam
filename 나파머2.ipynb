```python
import pandas as pd
star = pd.read_csv('6 class csv.csv')
star
```

```python
print(star['Star type'].unique())
print(star['Star type'].value_counts())
```

```python
print(star['Spectral Class'].unique())
print(star['Spectral Class'].value_counts())
```

```python
print(star['Star color'].unique())
print(star['Star color'].value_counts())
```

```python
star['Star color'] = star['Star color'].str.replace('-', ' ')
star['Star color'] = star['Star color'].str.rstrip()
star['Star color'] = star['Star color'].str.lower()
star['Star color'] = star['Star color'].str.replace('white yellow','yellow white')
print(star['Star color'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(star, hue='Star type', palette='RdBu',corner=True)
```

```python
p = sns.pairplot(star, hue = 'Star type', palette = 'RdBu', corner = True)
p._legend_data
```

```python
p = sns.pairplot(star,hue = 'Star type', palette = 'RdBu', corner = True)
handles = p._legend_data.values()
```

```python
labels = ['Brown Dwarf', 'Red Dwarf', 'White Dwarf', 'Main Sequence', 'Super Giant', 'Hyper Giant']
```

```python
p.legend.remove()
```

```python
p.fig.legend(title = '[Star type]', handles = handles, labels = labels, loc = 'center right', ncol = 1)
plt.show()
```
